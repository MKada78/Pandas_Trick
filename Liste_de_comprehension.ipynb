#### Liste de compréhension

Syntaxe:

[**Expression** for **élément** in **liste** if **condition**]

Exemple de liste de compréhension

```python
x = [i for i in range(10)]
x
```

1. Comparaison boucle **for** et **comprehension list**:

```python
#### Test boucle et liste de compréhension
squares = []

#### Pour chaque entier de 1 à 9
for x in range(10):
    #### On ajoute à la liste squares, le carré de l'entier
    squares.append(x**2)

#### On affiche la liste 
print(squares)
```

2. Générer des nombres

```python
#### Liste de compréhension

squares = [x**2 for x in range(10)]

print(squares)
```
