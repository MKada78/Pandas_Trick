```python
### Définition de la fonction et du dataframe de donnée
def regle(x,y):
    if x > 5 and y == "high":
        return 1 #### Return une condition qui se trouvera dans la colonne crée
    else:
        return 0

#### Construction d'un Data frame de test
df = pd.DataFrame({"col1": [10,45,98,25,55], "col2":["high", "low","high", "high","high"]})

df.head() #### Affichages des éléments du dataframe
```

```python
#### Application de la fonction lambda
### Apply est une fonction qui applique une fonction définit à 

df["new"] = df.apply(lambda x: regle(x['col1'], x['col2']), axis = 1) #### Crée une nouvelle colonne dans notre dataset
df.head()
df["babar"] = 0
```

```python
df.head()
```

```python
df["max"] = df[["col1", 'col2']].max(axis=1) #### Prendre le maximum de deux ou n colonnes et le mettre dans une nouvelle colonne du dataframe
```

```python
df.head()
```

```python
def multiplication_de_chiffre(x):
    """
        Cette fonction permet de multiplier les chiffres d'un nombre entre eux et en sort le résultat
            Entrée: Un nombre
            
            Sortie: Le produit des chiffres de ce nombre
    """
    chiffre = []
    chiffre = [int(j) for j in str(x)] #### Pour chaque éléments de la liste ["15"]
    #### On affiche le chiffre
    print(chiffre)
    
    #### On multiplie tout les éléments entre eux
    res = 1
    for i in chiffre:
        res = res *i
        
    return res

#### On lance notre fonction
multiplication_de_chiffre(99998999999)
```

```python
#### Application de la fonction à notre dataset

#### 1. Df.apply
#### 2. lambda x : fonction(x[colonne],...si plusieur colonne à comparer)
#### 3. axis = 1 pour la colonne

df["multiplication"] = df.apply(lambda x: multiplication_de_chiffre(x['col1']), axis = 1)
df.head() #### Yes
```

# Merge et Join

#### Merge avec inner join: Equivalent Inner Join

```python
pd.merge(customer, order, on ="customer_id", how="inner")
```

#### Merge avec un right join: Equivalent Right Join

```python
pd.merge(customer, order, on="customer_id", how="right")
```

#### Merge avec un left join: Equivalent Left Join SQL

```python
pd.merge(customer, order, on = "customer_id", how = "left")
```

#### Merge avec un suffixe aux colonnes dupliquées: Suffixe à ajouter au nom de la variable

```python
pd.merge(customer, order, on = "customer_id", how = "left", suffixes = ("_left", "_right")) #### Non aplicable ici
```

#### Merge basé sur les index: Basée sur l'index

```python
pd.merge(customer, order, right_index = True, left_index = True)
```
